```python
from glob import glob
import json
files = glob('*.json')

# ලැයිස්තුවක් වෙනුවට Dictionary එකක් භාවිතා කිරීම
unique_treatments = {}

for ff in files:
    with open(ff, 'r' , encoding='utf-8') as f:
        data = json.load(f)
    for i in data['treatments']:
        # 'treatment' එක key එක ලෙසත්, 'sinhala' එක value එක ලෙසත් එකතු කිරීම
        unique_treatments[i['treatment']] = i['sinhala']

# Alphabetical order එකට Keys අනුව sort කිරීම
sorted_unique_treatments = dict(sorted(unique_treatments.items()))

# ප්‍රතිඵලය බලාගැනීම සඳහා
print(json.dumps(sorted_unique_treatments, indent=4, ensure_ascii=False))
```
